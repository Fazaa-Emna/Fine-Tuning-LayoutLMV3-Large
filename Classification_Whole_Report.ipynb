```python
def create_bounding_box(bbox_data):
    xs = []
    ys = []
    for x, y in bbox_data:
        xs.append(x)
        ys.append(y)

    left = int(min(xs))
    top = int(min(ys))
    right = int(max(xs))
    bottom = int(max(ys))

    return [left, top, right, bottom]
```

```python
from pathlib import Path
import json
from easyocr import Reader

def process_image(image_path):
    # Convert image_path to a Path object
    image_path = Path(image_path)

    # Initialize OCR reader
    reader = Reader(['en'])

    # Perform OCR on the image
    ocr_result = reader.readtext(str(image_path), batch_size=16)

    ocr_page = []
    for bbox, word, confidence in ocr_result:
        ocr_page.append({
            "word": word, "bounding_box": create_bounding_box(bbox)
        })

    # Save OCR results to a JSON file
    with image_path.with_suffix(".json").open("w") as f:
        json.dump(ocr_page, f)

# Path to the folder containing images
images_folder = Path('/content/drive/MyDrive/MyDrive/MyFiles/Data/images')

# Iterate through all images in the folder
for image_file in images_folder.glob('*.png'):
    process_image(image_file)
```

```python
def scale_bounding_box(box: List[int], width_scale : float = 1.0, height_scale : float = 1.0) -> List[int]:
    return [
        int(box[0] * width_scale),
        int(box[1] * height_scale),
        int(box[2] * width_scale),
        int(box[3] * height_scale)
    ]
```

# **All in one fct: preprocessing + Classification**

```python
DEVICE = "cpu"
#"cuda:0" if torch.cuda.is_available() else
```

```python
def predict_document_image(
    image_path: Path,
    model: LayoutLMv3ForSequenceClassification,
    processor: LayoutLMv3Processor):

    json_path = image_path.with_suffix(".json")
    with json_path.open("r") as f:
        ocr_result = json.load(f)

        with Image.open(image_path).convert("RGB") as image:

            width, height = image.size
            width_scale = 1000 / width
            height_scale = 1000 / height

            words = []
            boxes = []
            for row in ocr_result:
                boxes.append(
                    scale_bounding_box(
                        row["bounding_box"],
                        width_scale,
                        height_scale
                    )
                )
                words.append(row["word"])

            encoding = processor(
                image,
                words,
                boxes=boxes,
                max_length=512,
                padding="max_length",
                truncation=True,
                return_tensors="pt"
            )

    with torch.inference_mode():
        output = model(
            input_ids=encoding["input_ids"].to(DEVICE),
            attention_mask=encoding["attention_mask"].to(DEVICE),
            bbox=encoding["bbox"].to(DEVICE),
            pixel_values=encoding["pixel_values"].to(DEVICE)
        )

    predicted_class = output.logits.argmax()
    return model.config.id2label[predicted_class.item()]
```

# **Perform Inference**
Prepare input tensors including token IDs for the text, attention mask, scaled bounding box coordinates, and possibly pixel values of the resized image.

```python
import os


# Assuming test_folder_path is the path to the folder containing images
test_folder_path = image_folder

# Get a list of all image files in the folder and sort them based on page number
image_files = [filename for filename in os.listdir(test_folder_path) if filename.endswith('.png')]
image_files.sort(key=lambda x: int(x.split('_')[1].split('.')[0]))  # Assuming filenames are in the format "page_6.png"

labels = []
predictions = []

# Iterate through each sorted image file
for filename in tqdm(image_files):
    image_path = Path(os.path.join(test_folder_path, filename))
    page_number = int(filename.split('_')[1].split('.')[0])  # Extract page number from filename
    labels.append(f"Page {page_number}")
    predictions.append(predict_document_image(image_path, model, processor))

# Now labels and predictions will contain the corresponding labels and predictions for each image in the folder
```

```python
# Print the predicted class along with the associated page number
for label, prediction in zip(labels, predictions):
    print(f"{label}: {prediction}")
```
